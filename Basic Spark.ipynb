```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Student profile").config("spark.some.config.option", "some-value") .getOrCreate()  
```

```python
marks1 = [('BDA', 88), ('ML', 92)]
marks2 = [('BDA', 91), ('ML', 85)]
marks3 = [('BDA',81)]
```

```python
Student = spark.createDataFrame([
                                 ('SU',1, marks1),
                                 ('AK',2,marks2),
                                 ('SM',3,marks3)
                                 ],
['Name', 'Rollno', 'Marks'] 
)
Student.show()
```

```python
data_collect = Student.collect()
s=0
c=0
for row in data_collect:
    x = row["Marks"]
    for i in x:
      if i[0]=='BDA':
        s=s+i[1]
      elif i[0]=='ML':
        c=c+i[1]
avg_BDA=s/3
avg_ML=c/2
print("Average marks of BDA: ")
print(avg_BDA)
print("Average marks of ML: ")
print(avg_ML)
```

```python
marks4 = [('BDA', 75)]
newStudent = spark.createDataFrame([('AM',4,marks4)],['Name','Rollno','Marks'])
newStudent.show()
```

```python
union = Student.union(newStudent)
union.show()
```
