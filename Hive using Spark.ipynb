```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Learning_Spark") \
    .getOrCreate()

sc = spark.sparkContext
```

# **Using Hive to create and read a table**

```python
from pyspark.sql import Row
from pyspark.sql import HiveContext
```

```python
sqlContext = HiveContext(sc)
```

```python
test_list = [('A', 25),('B', 20),('C', 25),('D', 18)]
rdd = sc.parallelize(test_list)
rdd
```

```python
people = rdd.map(lambda x: Row(name=x[0], age=int(x[1])))
schemaPeople = sqlContext.createDataFrame(people)
```

```python
sqlContext.registerDataFrameAsTable(schemaPeople, "test_table")
sqlContext.sql("show tables").show()
```

```python
sqlContext.sql("Select * from test_table").show()
```

```python
sqlContext.sql("Select name , age from test_table where age = 25").show()
```

```python
sqlContext.sql("Select max(age) from test_table").show()
```
